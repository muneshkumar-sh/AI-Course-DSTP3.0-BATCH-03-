```python
import numpy as np
```

```python
learning_rate=0.01 # step size
epochs=5 # number of iterations
X=np.array([1,2,3,4,5]) # input data
Y=np.array([3,4,2,5,6]) # target data
```

```python
m,b=0.0, 0.0 # initial parameters
n=len(X) # number of data points
```

```python
for i in range(epochs):
    Y_pred=m*X + b
    cost=((Y_pred - Y) ** 2).mean()
    # derivative helps in learning in small steps
    derivative_m=(-2/n) * sum(X* (Y-Y_pred)) # derivative wrt m
    derivative_b=(-2/n) * sum(Y-Y_pred) # derivative wrt b
    m -=learning_rate*derivative_m
    b -=learning_rate*derivative_b
    print(f"Epoch {i+1}/{epochs}, Cost: {cost}, m: {m}, b: {b}")
```
