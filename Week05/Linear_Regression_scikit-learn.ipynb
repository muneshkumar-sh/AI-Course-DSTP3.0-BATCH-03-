```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([[1], [3], [4], [5], [7]])
Y=np.array([3,5,7,9,11])
```

```python
# plot 1: scatter plot before regression
plt.figure(figsize=(6,4))
plt.scatter(X, Y, color='blue')
plt.title("Scaterr Plot before Linear Regression")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
model=LinearRegression()
model.fit(X, Y)
y_pred=model.predict(X)
print("Pridicted values: ", y_pred)
```

```python
# plot 2: scatter + fit line
plt.figure(figsize=(6,4))
plt.scatter(X, Y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label='Best Fit Line')
plt.title("Scaterr Plot with Regression best fit line")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()
```

```python
m=model.coef_
b=model.intercept_
```

```python
print("Slope (m): ", m)
print("Y Intercept (b): ", b)
```
