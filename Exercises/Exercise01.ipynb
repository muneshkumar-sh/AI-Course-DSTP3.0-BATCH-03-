```python
#3. Chat System using OOP concepts

class User:
    def __init__(self, name):
        self.name=name
        
class Message:
    def __init__(self, sender, text):
        self.sender=sender
        self.text=text

class ChatRoom:
    def __init__(self):
        self.users=[]
        self.history=[]

    def join(self, user):
        self.users.append(user)
        print(user.name, "joined the chat room.")

    def leave(self, user):
        self.users.remove(user)
        print(user.name, "left the chat room.")

    def send_message(self, user, text):
        msg=Message(user, text)
        self.history.append(msg)
        print(user.name, "sent:", text)

    def view_history(self):
        print("\nChat History:")
        for msg in self.history:
            print(msg.sender.name + ":", msg.text)
```

```python
# check above methods

#object
room=ChatRoom()

u1=User("Ali")
u2=User("Sara")

room.join(u1)
room.join(u2)

room.send_message(u1, "Hello everyone!")
room.send_message(u2, "Hi Ali!")

room.view_history()

room.leave(u2)
```
