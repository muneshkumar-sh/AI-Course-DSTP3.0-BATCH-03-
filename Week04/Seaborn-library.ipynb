```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("healthcare_patient_analytics_seaborn.csv")
```

```python
df = df.sample(500)

sns.scatterplot(
    data=df,
    x="treatment_cost",
    y="recovery_score",
    hue="visit_type"
)

plt.title("Treatment Cost vs Recovery Score")
plt.show()
```

```python
sns.relplot(
    data=df,
    x="treatment_cost",
    y="recovery_score",
    hue="visit_type",
    col="gender"
)
```

```python
sns.set_theme(style="whitegrid")
```

```python
# Matplotlib
plt.scatter(df["treatment_cost"], df["recovery_score"])

plt.xlabel("Treatment Cost")
plt.ylabel("Recovery Score")
```

```python
# Seaborn
sns.scatterplot(
    data=df,
    x="treatment_cost",
    y="recovery_score"
)
```
