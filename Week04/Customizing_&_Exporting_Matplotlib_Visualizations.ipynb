```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 50)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, label="sin(x)", color="b", linewidth=2)
ax.set_xticks(range(0, 11, 2))
plt.xticks(rotation=30)

ax.legend(loc="upper right")

ax.set_title("Customized Sine Wave Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.tight_layout()

plt.savefig("sine_plot.png", dpi=300, format="png")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "ecommerce_sales_analytics_5000.csv",
    parse_dates=["order_date"]
)
```

```python
fig, ax = plt.subplots()

ax.bar(df["payment_method"].value_counts().index, df["payment_method"].value_counts().values)

ax.set_xlabel("Payment Method")
ax.set_ylabel("Number of Orders")

ax.tick_params(axis="x", rotation=45)

plt.show()
```
