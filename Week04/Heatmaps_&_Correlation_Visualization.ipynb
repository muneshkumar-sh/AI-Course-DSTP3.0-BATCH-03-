```python
# Preparing the Healthcare Data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("healthcare_patient_analytics_seaborn.csv")

numeric_df = df[
    ["length_of_stay_days",
     "treatment_cost",
     "recovery_score",
     "readmission_risk"]
]
```

```python
# Correlation Matrix

corr_matrix = numeric_df.corr()

corr_matrix
```

```python
# Heatmap — Basic Visualization

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap (Healthcare Data)")

plt.show()
```

```python
# Masking Upper Triangle

import numpy as np

mask = np.triu(
    np.ones_like(corr_matrix, dtype=bool)
)

sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap (Masked)")

plt.show()
```
