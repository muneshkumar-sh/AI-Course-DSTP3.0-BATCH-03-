```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.set_title("Blank Figure Example")
ax.set_xlabel("X-axis Label")
ax.set_ylabel("Y-axis Label")

fig.savefig("blank_plot.png", dpi=100)

plt.show()
```

```python
# try on data  ----line plot
days = [1, 2, 3, 4, 5, 6, 7]
sales = [120, 150, 180, 170, 200, 220, 210]

plt.plot(days, sales,
         marker='o',
         linewidth=2,
         alpha=0.8,
         label="Daily Sales")

plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales Trend Over a Week")

plt.legend()
plt.show()
```

```python
# scatter plot

temperature = [20, 22, 25, 27, 30, 32, 35]
ice_cream_sales = [50, 65, 80, 90, 120, 140, 160]

plt.scatter(temperature, ice_cream_sales,
            alpha=0.7,
            marker='x',
            label="Sales vs Temp")

plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales")
plt.title("Relationship Between Temp & Sales")

plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "ecommerce_sales_analytics_5000.csv",
    parse_dates=["order_date"]
)

df = df.sample(n=100)
```

```python
monthly_revenue = (
    df
    .set_index("order_date")["revenue"]
    .resample("ME")
    .sum()
)

fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(monthly_revenue.index,
        monthly_revenue.values,
        linewidth=2)

ax.set_title("Monthly Revenue Trend")
ax.set_xlabel("Month")
ax.set_ylabel("Total Revenue")
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.scatter(df["unit_price"],
           df["revenue"],
           alpha=0.5)

ax.set_title("Unit Price vs Revenue")
ax.set_xlabel("Unit Price")
ax.set_ylabel("Revenue")
```
