```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("healthcare_patient_analytics_seaborn.csv")
```

```python
sns.histplot(
    data=df,
    x="treatment_cost",
    bins=30
)

plt.title("Distribution of Treatment Cost")

plt.show()
```

```python
sns.histplot(
    data=df,
    x="readmission_risk",
    hue="visit_type",
    multiple="stack",
    bins=30
)

plt.title("Re-admission by Visit Type")

plt.show()
```

```python
# KDEPlot — provides a smooth estimate of the distribution.

sns.kdeplot(
    data=df,
    x="recovery_score",
    hue="visit_type",
    fill=True
)

plt.title("KDE of Recovery Score by Visit Type")

plt.show()
```

```python
# ECDF — shows cumulative proportion.

sns.ecdfplot(
    data=df,
    x="length_of_stay_days",
    hue="visit_type"
)

plt.title("ECDF of Length of Stay Days")

plt.show()
```
