```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "ecommerce_sales_analytics_5000.csv",
    parse_dates=["order_date"]
)
```

```python
category_revenue = (
    df.groupby("product_category")["revenue"]
    .sum()
)

fig, ax = plt.subplots()

ax.bar(category_revenue.index,
       category_revenue.values)

ax.set_title("Total Revenue by Product Category")
ax.set_xlabel("Category")
ax.set_ylabel("Total Revenue")
```

```python
grouped = (
    df.groupby(["product_category", "region"])["revenue"]
    .sum()
    .unstack()   # set as unstack
)

grouped.plot(kind="bar", figsize=(10, 4))

plt.title("Revenue by Category and Region")
plt.ylabel("Revenue")

plt.show()
```

```python
grouped.plot(kind="bar",
             stacked=True,   # stack
             figsize=(8, 4))

plt.title("Stacked Revenue by Category and Region")
plt.ylabel("Revenue")

plt.show()
```
