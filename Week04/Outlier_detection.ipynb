```python
import pandas as pd
```

```python
df=pd.read_csv("ecommerce_sales_analytics_5000.csv", parse_dates=["order_date"])
```

```python
# IQR method
Q1=df["revenue"].quantile(0.25)
Q3=df["revenue"].quantile(0.75)
IQR=Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

outlier_iqr=df[(df['revenue']<lower_bound) | (df['revenue']>upper_bound)]
```

```python
outlier_iqr.shape
```

```python
outlier_iqr[["revenue", "product_category"]].head()
```

```python
df=df[(df['revenue']>=lower_bound) & (df['revenue']<=upper_bound)]
df.shape
```

```python
# z-score method
from scipy.stats import zscore

df = pd.read_csv("ecommerce_sales_analytics_5000.csv", parse_dates=["order_date"])

df["revenue_z"] = zscore(df["revenue"])

outliers_z = df[df["revenue_z"].abs() > 3]
```

```python
outliers_z.shape
```

```python
df = df[df["revenue_z"].abs() <= 3]

df.shape
```
