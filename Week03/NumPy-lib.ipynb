```python
# NumPy Random Numbers, Reproducibility
```

```python
# simulate 100k coin flips
rng=np.random.default_rng() # change mean
flips=rng.integers(0,2,size=100000) # 0=tails, 1=heads
prob_heads=np.mean(flips)
print("Estimated probability of heads: ", prob_heads)
```

```python
# simulate 100k coin flips
rng=np.random.default_rng(seed=40) # fix mean, generate same seq of random numbers 
flips=rng.integers(0,2,size=100000) # 0=tails, 1=heads
prob_heads=np.mean(flips)
print("Estimated probability of heads: ", prob_heads)
```

```python
# python loop
time_start=time.time()
squares=[i**2 for i in range(100000)]
time_end=time.time()
print("Loop time: ", time_end-time_start)

# numPy vectorized
time_start=time.time()
arr=np.arange(100000)
squares_np=arr**2
time_end=time.time()
print("Vectorized time: ", time_end-time_start)
```
