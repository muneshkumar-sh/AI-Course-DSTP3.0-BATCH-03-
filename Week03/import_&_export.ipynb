```python
import pandas as pd
import openpyxl
```

```python
df=pd.read_csv("Retail_sales.csv")

print(df.info())
```

```python
print(df)
```

```python
clean_df=df[df['Discount Percentage']>0]
subset=clean_df[["Store ID", "Product ID", "Date", "Units Sold", "Sales Revenue (USD)", "Discount Percentage", "Marketing Spend (USD)", "Store Location", "Product Category", "Day of the Week", "Holiday Effect"]]
```

```python
subset.to_excel("Subset_sales.xlsx", sheet_name='sales')
subset.to_json("Subset_sales.json", orient="records", date_format="iso")

print("File Exported")
subset.info()
```
