```python
import pandas as pd
```

```python
data={
    "Name": ["Ali", "Bob", "Charlie", "David"],
    "Age": [20, 34, 45, 30],
    "City": ["Karachi", "Sukkur", "Hyderabad", "Islamabad"],
    "Salary": [34000, 50000, 56000, 45000]
}

df=pd.DataFrame(data)
print("Original Dataset: \n", df)
```

```python
# labeled based selection (.loc)    
subset_loc=df.loc[1:3, ["Name", "Salary"]]     # ,both index include 1 & 3
print("Subset with loc: \n", subset_loc)
```

```python
# labeled based selection (.iloc)
subset_iloc=df.iloc[0:3, 1:4]   # ,index include 1 & 3 exclude
print("Subset with iloc: \n", subset_iloc)
```

```python
# boolean filtering
high_salary=df[df["Salary"]>50000]
print("High Salary Filter: \n", high_salary)
```

```python
# safe assignment using .loc
df.loc[df["Age"]<40, "Salary"]=65000
print("Low age salary change: \n", df.loc[df["Age"]<40, "Salary"])
```
