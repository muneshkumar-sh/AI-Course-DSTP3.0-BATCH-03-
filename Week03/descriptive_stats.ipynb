```python
df[["unit_price", "revenue"]].describe() # for revenue values
```

```python
# Quantiles
df['revenue'].quantile([0.25, 0.5, 0.75])
```

```python
# identify skewness
df['revenue'].mean(), df['revenue'].median()
```

```python
df['revenue'].skew()   # positive value => right skew
```

```python
# category level decritive statistics
df.groupby('product_category')['revenue'].describe()
```

```python
df.groupby('product_category')['revenue'].mean() # finding highest category
```

```python
Q1 = df["revenue"].quantile(0.25)
Q3 = df["revenue"].quantile(0.75)

IQR = Q3 - Q1

Q1, Q3, IQR
```
