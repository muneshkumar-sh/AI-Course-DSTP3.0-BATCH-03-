```python
import pandas as pd
```

```python
df=pd.read_csv("ecommerce_sales_analytics_5000.csv", parse_dates=['order_date'])
```

```python
# setting dateTime index
df=df.set_index('order_date')
```

```python
df.index
```

```python
# Monthly Revenue Analysis
monthly_revenue = df["revenue"].resample("ME").sum()

monthly_revenue
```

```python
monthly_revenue.pct_change()  # trend interpretation
```

```python
monthly_revenue.index.is_monotonic_increasing   # verify
```

```python
# check
print(monthly_revenue.sum(), df["revenue"].sum())
monthly_revenue.sum()<=df["revenue"].sum()
```
