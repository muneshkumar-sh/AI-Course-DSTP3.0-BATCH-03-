```python
import pandas as pd
```

```python
# Sample DataFrames
customers = pd.DataFrame({
    'CustomerID': [1,2,3],
    'Name': ['Alice','Bob','Charlie']
})
```

```python
orders = pd.DataFrame({
    'OrderID': [101,102,103,104],
    'CustomerID': [1,2,2,3],
    'Amount': [250,150,200,300]
})
```

```python
# inner join on Costumer ID
df_inner=pd.merge(customers, orders, on='CustomerID', how='inner')
print("Inner Merge: \n", df_inner)
```

```python
# left join on Costumer ID
df_left=pd.merge(customers, orders, on='CustomerID', how='left')
print("left Merge: \n", df_left)
```

```python
# right join on Costumer ID
df_right=pd.merge(customers, orders, on='CustomerID', how='right')
print("Right Merge: \n", df_right)
```

```python
# outer join on Costumer ID
df_outer=pd.merge(customers, orders, on='CustomerID', how='outer')
print("Right Merge: \n", df_outer)
```

```python
# concatenate row example
df_concat=pd.concat([customers, orders])
print("Concatenate Data: \n", df_concat)
```
