```python
import pandas as pd
```

```python
df=pd.read_csv("ecommerce_sales_analytics_5000.csv", parse_dates=['order_date'])
```

```python
df.info()
```

```python
# map() - label hight value orders
df['payment_type']=df['payment_method'].map({
    "Card": "Digital",
    "Wallet": "Digital",
    "COD": "Cash"
})
```

```python
# verify
df['payment_type'].unique()
```

```python
df[['payment_method', 'payment_type']].head()
```

```python
df.groupby('payment_type').size()
```

```python
# apply() - Create revenue category

def revenue_category(row):
    if row["revenue"] > 500:
        return "High"
    elif row["revenue"] >= 200:
        return "Medium"
    else:
        return "Low"

df["revenue_category"] = df.apply(revenue_category, axis=1)
```

```python
#verify
df['revenue_category'].value_counts()
```

```python
# validate logic manually
df[['revenue', 'revenue_category']].sample(10)
```

```python
# pipe() - Method chaining

def add_profit(data):
    data["profit"] = data["revenue"] * 0.20
    return data

def add_net_revenue(data):
    data["net_revenue"] = data["revenue"] - data["profit"]
    return data

df = (
    df
    .pipe(add_profit)
    .pipe(add_net_revenue)
)
```
