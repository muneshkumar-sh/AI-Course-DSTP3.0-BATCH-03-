```python
import pandas as pd
```

```python
df=pd.DataFrame({
    'Category': [" Electronics ", "furniture", "Home-Appliance", "electronics ", "FURNITURE "]
})

print(df)
```

```python
# strip spaces and convert to lower case
df['category_clean']=df['Category'].str.strip().str.lower()
print(df)
```

```python
# replace hyphons with space
df['category_clean']=df['category_clean'].str.replace('-', ' ')
print(df)
```
