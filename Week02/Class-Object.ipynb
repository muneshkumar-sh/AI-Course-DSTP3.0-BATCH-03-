```python
class Car:
    def __init__(self, color):
        self.color=color

    def start(self):
        print(f"{self.color} car started")

# 1st object 
my_car=Car("Red")
my_car.start()

# 1st object 
my_brother_car=Car("Silver")
my_brother_car.start()
```

```python
class Car:
    def drive(self):
        print("The car is moving")

# object 
car1=Car()
car1.drive()
```

```python
class Car:
    color="Red"
    def drive(self):
        print(f"{self.color} Car is moving")

# object 
car1=Car()
car1.drive()
```

```python
class Car:
    color="Red"
    def drive(self):
        print(f"{self.color} Car is moving")
    # change defalt color
    def setColor(self, new_color):
        self.color=new_color

# object 
car2=Car()
car2.setColor("Green")
car2.drive()
```
