```python
class Dog:
    def speak(self):
        print("Woof!")
```

```python
class Cat:
    def speak(self):
        print("Meow!")
```

```python
# call both methods
for pet in [Dog(), Cat()]:
    pet.speak()
```

```python
# methods Overriding
class Animal:
    def speak(self):
        print("Some sound")
class Dog(Animal):
    def speak(self):
        print("woof! woof!")
        
childDog=Dog()        
childDog.speak() # override parent class method
```

```python
# method overloading
class Calculator:
    def add(self, a, b=0, c=0): # set b & c default values if no one argument is passed for b & c , but here a is compulsory for add 
        return a+b+c
```

```python
# object
calc=Calculator()
```

```python
# one numbers
print(calc.add(10)) # b & c are 0's
# add two numbers
print(calc.add(10,5)) # c is 0
# add three numbers
print(calc.add(10,12,15))
```
