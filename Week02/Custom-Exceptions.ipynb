```python
class InvalidAgeError(Exception):
    pass
```

```python
age=int(input("Enter Age: "))
if age<0:
    raise InvalidAgeError(f"Age can not be neagtive!, {age}")
```

```python
try:
    age=int(input("Enter Age: "))
    if age<0:
        raise InvalidAgeError(f"Age can not be neagtive!, {age}")
except InvalidAgeError as e:
    print(e)
```
