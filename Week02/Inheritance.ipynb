```python
class Animal:
    def speak(self):
        print("Some sound")
```

```python
class Dog(Animal):   # Dog class inherit from Animal
    pass   # pass keyword shows that still in this we can not define anything
```

```python
# object
dog1=Dog()
```

```python
dog1.speak()
```

```python
class Dog(Animal):
    def bark(self):
        print("woof!")
```

```python
childDog=Dog()
```

```python
# can access both parent & child methods
childDog.bark()
childDog.speak()
```

```python
# methods Overriding
class Animal:
    def speak(self):
        print("Some sound")
class Dog(Animal):
    def speak(self):
        print("woof! woof!")
        
childDog=Dog()        
childDog.speak() # override parent class method
```
