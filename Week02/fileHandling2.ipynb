```python
# to handle this type of error
try:
    with open("first.txt", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("File not found")
```

```python
filename="my_info.txt"
name="GitHub copilot"
age=1
```

```python
# try to read exsisting file , if missing then create it with name and age & read
try:
    with open(filename, "r") as fin:
        print(fin.read())
except FileNotFoundError:
    with open(filename, "w") as fout:
        fout.write(f"{name}\n{age}\n")
    with open(filename, "r") as fin:
        print(fin.read())
```
