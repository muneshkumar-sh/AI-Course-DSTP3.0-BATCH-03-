```python
class Dog:
    def __init__(self):
        self.name="Tony"

    def bark(self):
        print(f"{self.name} says woof!")

dog1=Dog()
dog1.bark()
```

```python
class Car:
    def __init__(self, brand, color):
        self.brand=brand
        self.color=color

car1=Car("Toyota", "Silver")
print(car1.brand)
print(car1.color)
```

```python
car1.color="Blue"
car2=Car("Toyota", "Silver")
print(car1.color)
print(car2.color)
```
