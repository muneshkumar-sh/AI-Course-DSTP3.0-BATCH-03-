```python
# create dictionary
student={
	"name": "Munesh",
	"age": 21,
	"course": "Python"
}

print(student)
```

```python
# access values
print(student["name"])
print(student.get("age"))
```

```python
# methods
student["age"]=22 # update
student["gpa"]=4.0 # add
print(student)
```

```python
student["gpa"]=4.0 # add
print(student)
```

```python
print(student.pop("gpa"))
```

```python
# looping dictionay
for keys, values in student.items():
    print(keys , " : " , values)
```
