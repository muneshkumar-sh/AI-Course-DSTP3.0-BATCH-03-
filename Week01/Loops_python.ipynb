```python
# nested loop

i=4
rows=i
cols=i
matrix=[]

for r in range(rows):
    row=[]
    for c in range(cols):
        row.append(0)
    matrix.append(row)

for row in matrix:
    print(" ".join(map(str, row)))
```

```python
for level in range(1, rows+1):
    print(' ' * (rows-level) + '*' * (2*level-1))
```
