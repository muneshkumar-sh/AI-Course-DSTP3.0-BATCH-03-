```python
score=int(input("Enter Score: "))
```

```python
# nested if

if score>=85:
    print("Achieved Grade A")
elif score>=70 and score <85:  # combine conditions
    print("Achieved Grade B")
elif score>=50 and score <70:
    print("Achieved Grade C")
elif score<50:
    print("Achieved Grade F")
```

```python
age=int(input("Enter Age: "))
if age>=17:
    country=input("Enter Country Name: ")
    if country=="Pakistan":
        print("You can vote in Pakistan")
    else:
        print("You can not vote in pakistan")
else: 
    print("You are not Eligible")
```

```python
age=int(input("Enter Age: "))
country=input("Enter Country Name: ").capitalize() # capitalize() use to make first letter capital
if age>=17 and country=="Pakistan":   # combine Conditions
    print("You are eligible for Vote")
else:
    print("You are not eligible for Vote")
```
