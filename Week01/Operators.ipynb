```python
print(5+3) # addition
print(10-4) # substraction
```

```python
print(5*10) # multiplication
print(5/10) # division
```

```python
print(10//4) # floor divide
```

```python
print(10%3) # remainder
print(2**5) # exponent
```

```python
# equality
print(2==3)
print(5==5)
```

```python
# inequality
print(2!=3)
print(2!=2)
```

```python
# greater than
print(3>5)
print(5>1)
print(2>=2)
```

```python
# less than
print(3<5)
print(5<1)
print(2<=2)
```

```python
# binary operators
print(True and False) # binary 'and' operator
print(True or False) # binary 'or' operator
print(not True) # binary 'not' operator
```

```python
# perform arithmetic operators
x=10
print(x)
x+=3 # addition x+3=13
print(x) # 13
x-=5 # subtraction x-5=8
print(x) # 8
x*=5 # multiplication x*5=40
print(x) # 40
x/=5 # dividion x/5=4.0
print(x) # 4.0
```
